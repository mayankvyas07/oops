```python
Q1. Explain Class and Object with respect to Object-Oriented Programming. Give a suitable example.

Class: A class is a blueprint or template for creating objects. It defines the properties (attributes) and behaviors (methods) that objects of that class will have. It acts as a blueprint from which you can create multiple objects with similar characteristics.
Object: An object is an instance of a class. It is a concrete entity that embodies the characteristics and behaviors defined by the class. Objects are the actual data structures that hold the data and provide methods to operate on that data.
Example:
Let's consider a class Car:

python
Copy code
class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def start_engine(self):
        print(f"The {self.make} {self.model}'s engine is starting.")

# Creating objects from the class
car1 = Car("Toyota", "Corolla")
car2 = Car("Ford", "Mustang")

# Using object methods
car1.start_engine()
car2.start_engine()
Q2. Name the four pillars of OOPs:
The four pillars of object-oriented programming are:

Encapsulation: Encapsulation refers to the bundling of data (attributes) and methods (functions) that operate on the data into a single unit called a class. It restricts direct access to some of the object's components and hides the internal details.
Abstraction: Abstraction involves simplifying complex reality by modeling classes appropriate to the problem and neglecting irrelevant details. It provides a clear separation between what an object does and how it's accomplished.
Inheritance: Inheritance allows a new class (subclass or derived class) to inherit properties and behaviors (attributes and methods) from an existing class (base class or superclass). This promotes code reuse and hierarchy among classes.
Polymorphism: Polymorphism allows objects of different classes to be treated as objects of a common superclass. It enables objects to be processed in a general way, but different subclasses can provide their own implementations of methods.
Q3. Explain why the init() function is used. Give a suitable example:
The __init__() function is a special method (constructor) in a class that is automatically called when an object of that class is created. It initializes the attributes of the object. It's used to set up the initial state of an object by assigning values to its attributes.

Example:
Continuing from the previous car example:

python
Copy code
class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model

car1 = Car("Toyota", "Corolla")
print(car1.make)  # Output: Toyota
print(car1.model)  # Output: Corolla
Q4. Why self is used in OOPs?
In object-oriented programming, self is a reference to the instance of the class. It is used to access the attributes and methods of the instance within its own methods. It allows you to differentiate between instance variables (attributes) and local variables within methods.

Q5. What is inheritance? Give an example for each type of inheritance:
Inheritance is a fundamental concept in OOP that allows a new class (subclass or derived class) to inherit properties and behaviors (attributes and methods) from an existing class (base class or superclass). This promotes code reuse and allows for creating a hierarchy of related classes.

Types of Inheritance:

Single Inheritance: A subclass inherits from a single superclass.
python
Copy code
class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        print("Woof!")

class Cat(Animal):
    def speak(self):
        print("Meow!")
Multiple Inheritance: A subclass inherits from multiple superclasses.
python
Copy code
class A:
    def method_a(self):
        pass

class B:
    def method_b(self):
        pass

class C(A, B):
    def method_c(self):
        pass
Multilevel Inheritance: One class inherits from another class, which in turn inherits from another class.
python
Copy code
class A:
    pass

class B(A):
    pass

class C(B):
    pass
Hierarchical Inheritance: Multiple subclasses inherit from a single superclass.
python
Copy code
class Shape:
    def area(self):
        pass

class Circle(Shape):
    def area(self):
        pass

class Square(Shape):
    def area(self):
        pass
Remember, the choice of inheritance type depends on the problem you are solving and the relationships between different classes in your application.
```
